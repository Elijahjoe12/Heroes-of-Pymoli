```python
# Importing necessary library
import pandas as pd
```

```python
purchase_data = pd.read_csv('purchase_data.csv')
purchase_data.head()
```

Player Count

Total Number of Players

```python
# Total number of players
players = purchase_data["SN"].nunique()
players
```

Purchasing Analysis (Total)

Number of Unique Items

Average Purchase Price
Total Number of Purchases
Total Revenue
Create a summary data frame to hold the results

```python
# Number of unique items
items = purchase_data["Item ID"].nunique()

# Average Purchase Price
avg_pur = purchase_data["Price"].mean()

# Total Number of Purchases
tot_purchase = purchase_data["Purchase ID"].nunique()

# Total Revenue
tot_rev = purchase_data["Price"].sum()

purchase = [{"Unique Items": items,
             "Average Purchase Price": avg_pur,
             "Total Purchases": tot_purchase,
             "Total Revenue": tot_rev}]
purchase_summary = pd.DataFrame(purchase)

# formatting
purchase_summary["Average Purchase Price"] = purchase_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_summary["Total Revenue"] = purchase_summary["Total Revenue"].map("${:,.2f}".format)
purchase_summary
```

Gender Demographics

Percentage and Count of Male Players

Percentage and Count of Female Players

Percentage and Count of Other / Non-Disclosed

Create a summary data frame to hold the results

```python
# Group by the number of unique SN in genders
gender_count = purchase_data.groupby("Gender")['SN'].nunique()

# percentage of male, female, and other players
percent_gender = (gender_count/players)*100


gender_summary = pd.DataFrame({"Number of Players": gender_count,
                               "Percentage of Players":percent_gender })

# formatting
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:,.2f}%".format)

gender_summary
```

Purchasing Analysis (Gender)

The below each broken by gender

Purchase Count

Average Purchase Price

Total Purchase Value

Average Purchase Total per Person by Gender
    
Create a summary data frame to hold the results

```python
# Purchase Count
purchase_count = purchase_data.groupby('Gender')["Price"].count()

# Average Purchase Price
avg_pur_gen = purchase_data.groupby("Gender")["Price"].mean()

# Total Purchase Value
tot_pur_gen = purchase_data.groupby("Gender")["Price"].sum()

# Average Purchase Total per Person by Gender
avg_pur_tot = tot_pur_gen/gender_count

purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                  "Average Purchase Price": avg_pur_gen,
                                  "Total Purchase Value": tot_pur_gen,
                                  "Average Purchase Total per Person": avg_pur_tot})
# formatting
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis["Average Purchase Total per Person"] = purchase_analysis["Average Purchase Total per Person"].map("${:,.2f}".format)

purchase_analysis

```

Age Demographics

The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

Purchase Count

Average Purchase Price

Total Purchase Value

Average Purchase Total per Person by Age Group

Create a summary data frame to hold the results

```python
# setting bin
age_bins = [0, 9, 14, 18, 23, 27, 31, 35, 39, 80]
age_labels = ["Under 10", "10-14", "15-18","19-23", "24-27", "28-31", "32-35", "36-39", "40 and above"]

age_demo = purchase_data.copy()

age_demo["Age Group"] = pd.cut(age_demo["Age"], bins=age_bins, labels=age_labels)

# convert new column to string
age_demo["Age Group"] = age_demo["Age Group"].astype(str)

# Purchase Count
age_pur_count = age_demo.groupby("Age Group")["Price"].count()

# Average Purchase Price
age_avg_pur = age_demo.groupby("Age Group")["Price"].mean()

# Total Purchase Value
age_tot_pur = age_demo.groupby("Age Group")["Price"].sum()

# Average Purchase Total per Person by Age Group
age_count = age_demo.groupby("Age Group")['SN'].nunique()
age_avg_tot = age_tot_pur/age_count

agedemo_summary = pd.DataFrame({"Purchase Count": age_pur_count,
                                "Average Purchase Price": age_avg_pur,
                                "Total Purchase Value": age_tot_pur,
                                "Average Purchase Total per Person": age_avg_tot})

# formatting
agedemo_summary["Average Purchase Price"] = agedemo_summary["Average Purchase Price"].map("${:,.2f}".format)
agedemo_summary["Total Purchase Value"] = agedemo_summary["Total Purchase Value"].map("${:,.2f}".format)
agedemo_summary["Average Purchase Total per Person"] = agedemo_summary["Average Purchase Total per Person"].map("${:,.2f}".format)

agedemo_summary
```

Top Spenders

Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

SN

Purchase Count

Average Purchase Price

Total Purchase Value

```python
# Purchase Count
sn_pur_count = purchase_data.groupby("SN")['Purchase ID'].count()

# Average Purchase Price
sn_avg_pur =  purchase_data.groupby("SN")["Price"].mean()

# Total Purchase Value
sn_tot_pur = purchase_data.groupby("SN")["Price"].sum()

spenders = pd.DataFrame({"Purchase Count": sn_pur_count,
                         "Average Purchase Price": sn_avg_pur,
                         "Total Purchase Value": sn_tot_pur})

# getting the top spenders 
top_spenders = spenders.sort_values(["Total Purchase Value"], ascending=False)

#formatting
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:,.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:,.2f}".format)

top_spenders.head()
```

Most Popular Items

Identify the 5 most popular items by purchase count, then list (in a table):

Item ID

Item Name

Purchase Count

Item Price

Total Purchase Value

```python
# Item Name
itm_name = purchase_data.groupby("Item ID")['Item Name'].unique()

# Purchase Count
itm_pur = purchase_data.groupby("Item ID")['Purchase ID'].count()

 # Total Purchase Value
itm_tot  = purchase_data.groupby("Item ID")['Price'].sum()

# Item price
itm_price  = itm_tot/itm_pur

item_summary = pd.DataFrame({"Item Name": itm_name,
                             "Purchase Count": itm_pur,
                         "Item Price": itm_price,
                         "Total Purchase Value": itm_tot})

# sorting
pop_itm = item_summary.sort_values('Purchase Count', ascending=False)

# formatting
pop_itm["Item Price"] = pop_itm["Item Price"].map("${:,.2f}".format)
pop_itm["Total Purchase Value"] = pop_itm["Total Purchase Value"].map("${:,.2f}".format)

pop_itm.head()
```

Most Profitable Items

Identify the 5 most profitable items by total purchase value, then list (in a table):

Item ID

Item Name

Purchase Count

Item Price

Total Purchase Value

```python
# Sorting for most popular items
prof_itm = item_summary.sort_values('Total Purchase Value', ascending=False)

# formatting
prof_itm["Item Price"] = prof_itm["Item Price"].map("${:,.2f}".format)
prof_itm["Total Purchase Value"] = prof_itm["Total Purchase Value"].map("${:,.2f}".format)

prof_itm.head()
```
